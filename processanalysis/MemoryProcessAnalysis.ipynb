# Memory Process Analysis
Lets look at the memory we just collected. In volatility3 you no longer have to identify the profile with 'imageinfo' it will scan for the profile to use and download the required symbols to do the analysis. Lets look at the plugins available with the '-h' option

```powershell
python ../volatility3/vol.py -h
```

```powershell
python ../volatility3/vol.py -f E:\memoryimage.raw windows.info.Info
```

Now that we have seen the basic information from this memory dump we can start analysis. The first analysis to focus on is the running processes. There are several ways to look at running processes. The first way we will look at running processes is with `pslist`. `pslist` analysis is similar to executing pslist from Sysinternals and it looks at the _EPROCESS linked list data structure in memory. The '-q' flag will supress the progress output.

```powershell
python ../volatility3/vol.py -q -f E:\memoryimage.raw windows.pslist.PsList
```

```powershell
python ../volatility3/vol.py -q -r csv -f E:\memoryimage.raw windows.pslist.PsList
```

```powershell
python ../volatility3/vol.py -q -r csv -f E:\memoryimage.raw windows.pslist.PsList > plist.csv
```

```powershell
conda install csvkit
```

```powershell
csvlook -h
```

```powershell
csvlook plist.csv
```

```powershell
csvcut -c 2,3,4,6,7,10,11 plist.csv | csvlook
```

It is sometimes easier to review processes as a tree structure instead of a flat list. The `pstree` command lists the process list as a tree structure with the trunks being the parent process and branches the spawned process. Execute the following command to view the process tree and review if all the critical processes are there and if anything looks supicous based on reviewing common processes

```powershell
python ../volatility3/vol.py -q -f E:\memoryimage.raw windows.pstree.PsTree 
```

The other process analysis is `psscan` which carves the EPROCESS structure from memory and does not rely on a linked data struture like `pslist`. Go ahead and execute the command to list what processes are in memory.

```powershell
python ../volatility3/vol.py -q -f E:\memoryimage.raw windows.psscan.PsScan
```
