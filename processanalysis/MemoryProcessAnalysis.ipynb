# Memory Process Analysis
Lets look at the memory we just collected. To start you have to identify the memory image profile of the collected memory image. In this case we know what system we collected it on but in some cases you may just be procided with a raw image and may not know what system it was collected on. 

```powershell
python ../volatility/vol.py -f ..\memorycollection\memoryraw.dd imageinfo
```

Now that we have identified the profile to use we can specificy that flag and begin analysis. The first analysis to focus on is the running processes. There are several ways to look at running processes. The first way we will look at running processes is with `pslist`. `pslist` analysis is similar to executing pslist from Sysinternals and it looks at the _EPROCESS linked list data structure in memory.

```powershell
python ../volatility/vol.py -f C:\memoryraw --profile=Win10x64_14393 pslist
```

It is sometimes easier to review processes as a tree structure instead of a flat list. The `pstree` command lists the process list as a tree structure with the trunks being the parent process and branches the spawned process. Execute the following command to view the process tree and review if all the critical processes are there and if anything looks supicous based on reviewing common processes

```powershell
python ../volatility/vol.py -f C:\memoryraw --profile=Win10x64_14393 pstree
```

The other process analysis is `psscan` which carves the EPROCESS structure from memory and does not rely on a linked data struture like `pslist`. Go ahead and execute the command to list what processes are in memory.

```powershell
python ../volatility/vol.py -f C:\memoryraw --profile=Win10x64_14393 psscan
```

It can be troublesome to compare all the process enumeration capabiliteis in volatility so reviewing them all at the same time can be done with the command `psxview` 

```powershell
python ../volatility/vol.py -f C:\memoryraw --profile=Win10x64_14393 psxview
```

As you can see from the output there is several process listing enumeration capabilities from memory analysis. The list of the capabiliteis are below:

* pslist 
* psscan 
* thrdproc 
* pspcid 
* csrss 
* session
* deskthrd

We have already discussed `pslist` and `psscan` process enumeration. The other four process enumerations capabilities are detailed below:

* thdproc - This process enumeration capability that looks for the thread data structure and assocaited _EPROCESS all processes have to have a least a single thread so if a malious process is hiding a _EPROCESS structure it would also have to hide all the thread data structures as welll

* pspcid - This process enumeration capability that looks for another handle reference data storage structure in the kernel called the PspCid handle table. 

* csrss - This process enumeration capability looks processes that are not associated with the `csrss` process. All processes are assoicated with this process and if they do not have handles with this process they should be supicious unless they it is it the `csrss` process itself or a parent process of `csrss`

* session - This process enumeration capability looks at the SessionProcessLinks member of _EPROCESS that associates all processes belonging to a particular user’s logon session. All processes have a user and a supicious process would have to unlink all these data structures to hide.

* deskthrd - Another data structure that identifies processes is the Desktop (tagDESKTOP) data structure. These structures store a list of all threads attached to each desktop, and you can easily map a thread back to its owning process.
