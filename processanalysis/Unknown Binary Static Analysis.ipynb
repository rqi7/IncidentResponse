### Hashes

Hashes help identify a file indistigusable from others. Common hash types you will identify in digital forensics and incident reponse are below:

 * MD5
 * SHA1
 * SHA256
 * SSDEEP

These hashes can used used to search malware data bases such as VirusTotal. Below are some command lines to generate hashes for binaries we have identified while doing memory analysis.

```powershell
md5sum .\dump\pid.796.irykmmww.dll.0x292e60.0x150000.dmp
```

```powershell
Get-FileHash  .\dump\pid.796.irykmmww.dll.0x292e60.0x150000.dmp -Algorithm MD5
```

```powershell
Get-FileHash  .\dump\pid.796.irykmmww.dll.0x292e60.0x150000.dmp -Algorithm SHA1
```

```powershell
Get-FileHash  .\dump\pid.796.irykmmww.dll.0x292e60.0x150000.dmp -Algorithm SHA256
```

Remember that hashes from files extracted from memory might not match what is on disk.
### Strings

Strings are useful to help us identify basic functionality or possible IOCs from files or memory space. You can use 'strings' but the Mandiant FLARE team has created a tool called FLOSS which iterates through all string possibilities and does some dynamic decoding of strings. Below is an example of executing FLOSS on a binary we extracted:

```powershell
floss .\dump\pid.796.irykmmww.dll.0x292e60.0x150000.dmp
```

To get a better since of the unknown binary extracted you can use the list of tools below to identify if they are signed and what APIs they might use or other resources they relay on to derive functionality or malicousness possibility:
* PEID
* PEView
* PE Explorer
* CFF Explorer
* sigcheck

Most of these tools are GUI driven but sigcheck executing is shown below:

```powershell
sigcheck -v .\dump\pid.796.irykmmww.dll.0x292e60.0x150000.dmp
```

CAPA is another tool developed by Mandiant's FLARE team which uses string signatures to better identify unknown binary capabilies just from strings. Below is an example use of the tool from a binary extracted from memory:

```powershell
capa .\dump\pid.796.irykmmww.dll.0x292e60.0x150000.dmp
```
